```python
#opening input file
with open('day5-input.txt', 'r') as f:
    lines = f.readlines()
    print(lines)
```

```python
lines.index("\n")
```

```python
#defining the stack part
stack = lines[:lines.index("\n")]
stack
```

```python
#replace empty places with a placeholder ZZZ
stackReplaced = [entry.replace("    "," ZZZ ").strip() for entry in stack]
stackReplaced[len(stackReplaced)-1] = stackReplaced[len(stackReplaced)-1].replace("   "," ")
print(stackReplaced)
```

```python
stackItemsUnordered = [entry.split(" ") for entry in stackReplaced]
print(stackItemsUnordered)
```

```python
#remove empty array entries.
for entry in stackItemsUnordered:
    while "" in entry:
        entry.remove("")
stackItemsUnordered
```

```python
noOfStacks = int(stackItemsUnordered[-1][-1])
noOfStacks
```

```python
#alining all stacks into one array
stackItemsOrdered = []
for i in range(noOfStacks):
    array = []
    for j in range(len(stackItemsUnordered)-1,-1,-1):
        array.append(stackItemsUnordered[j][i])
    stackItemsOrdered.append(array)
    
#removing the placeholder items
for stackItems in stackItemsOrdered:
    while "ZZZ" in stackItems:
        stackItems.remove("ZZZ")
stackItemsOrdered
```

```python
#defining commands part
commands = lines[lines.index("\n")+1:]
commands
```

```python
commands = [entry.strip() for entry in commands]
commands
```

```python
#making a list out of every command
commandsSplit = [entry.split(" ") for entry in commands]
commandsSplit
```

```python
#changing every command entry into a dictionary
commandsDict = []
for i in range(len(commandsSplit)):
    dictionary = {"move":int(commandsSplit[i][1]),
                 "from":int(commandsSplit[i][3]),
                 "to":int(commandsSplit[i][5]),                 
                 }
    commandsDict.append(dictionary)
commandsDict       
```

```python
#function move only one item at a time (crane 9000)
def moveItems(fromNo,toNo):
    listToTake = stackItemsOrdered[fromNo-1]
    listToAdd = stackItemsOrdered[toNo-1]
    item = listToTake.pop()
    listToAdd.append(item)       
```

```python
#function to move all items at once (crane 9001)
def moveAllItems(noOfItems,fromNo,toNo):
    listToTake = stackItemsOrdered[fromNo-1]
    listToAdd = stackItemsOrdered[toNo-1]
    items = listToTake[len(listToTake)-noOfItems:]
    listToAdd.extend(items) 
    i = 0
    for i in range(noOfItems):
        listToTake.pop()
        i+=1
```
